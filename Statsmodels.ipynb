```python
def bootstrap_linreg(x, y, n_bootstrap):
    x = sm.add_constant(x)
    model = sm.OLS(y, x).fit()
    coef = model.params
    
    coef_boot = pd.DataFrame([], columns = x.columns)
    for i in range(n_bootstrap):
        x_boot = resample(x, n_samples = data.shape[0])
        y_boot = y[x_boot.index] # 
        
        model_boot = sm.OLS(y_boot, x_boot).fit()
        coef_boot = coef_boot.append(model_boot.params, ignore_index = True)
    
    mean_boot = coef_boot.mean()
    se_boot = coef_boot.std()
    bias = coef - mean_boot
    
    return pd.Series([coef, coef_boot, mean_boot, se_boot, bias], 
                     index = ["coef", "coef_boot", "mean_boot", "se_boot", "bias"])
```

```python
boot = bootstrap_linreg(x, y, n_bootstrap = 100)
```

```python
x = sm.add_constant(x)
```

```python
alpha = [1, 0.3, 0.1, 0.03, 0.01, 0.003, 0.001, 
         0.0003, 0.0001, 0.00003, 0.00001, 0.000003, 0.000001]
k = 10
folds = np.random.randint(0,k, data.shape[0])
cv_error = np.ones((k, len(alpha)))
```

```python
for i in range(len(alpha)):
    for j in range(k):
        trainIndices = x[folds != j].index.values.astype(int)
        testIndices = x[folds == j].index.values.astype(int)

        lasso_model = sm.OLS(y[trainIndices], x.loc[trainIndices,]).fit_regularized(alpha = alpha[i], L1_wt = 1)
        ypred = lasso_model.predict(x.loc[testIndices,])
        
        cv_error[j, i] = np.mean((y[testIndices]-ypred)**2)
```

```python
cv_error = pd.DataFrame(cv_error, columns = alpha)
```
